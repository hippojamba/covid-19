```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

# Read csv to DataFrame
infected = pd.read_csv('sweden-infected-by-region.csv')
```

```python
# Select Halland and Statistikdatum column for our plot
date_of_first_registered_case = '3/6/2020'
halland = infected[infected['Statistikdatum'] > date_of_first_registered_case].loc[ : , 'Halland']
date = infected[infected['Statistikdatum'] > date_of_first_registered_case].loc[ : , 'Statistikdatum']
```

```python
plt.style.use('fivethirtyeight') # Set theme of plot


fig, ax = plt.subplots(figsize=(50, 10))

# Plot and add labels
ax.plot(date, halland)

ax.set(title='Nya registrerade fall i Halland',
       xlabel='Datum',
       ylabel='Antal nya fall');

ax.axhline(y=halland.mean(), 
            color='r', 
            linestyle='--', 
            label="Average");
```

```python
# Save figure to .png
fig.savefig('halland-infected-by-date.png', dpi=100)
```
